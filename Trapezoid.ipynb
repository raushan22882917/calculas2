```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to be integrated
def f(x):
    return x**2 + 2*x + 1

# Define the exact integral of the function for error calculation
def exact_integral(a, b):
    return (b**3 - a**3) / 3 + (b**2 - a**2) + (b - a)

# Define the limits of integration
a = 0  # lower limit
b = 2  # upper limit

# Define the maximum number of intervals
n_max = 100

# Initialize arrays to store the number of intervals and corresponding errors
n_values = np.arange(1, n_max + 1)
errors = []

# Loop through different number of intervals
for n in n_values:
    # Calculate the width of each interval
    h = (b - a) / n

    # Generate an array of equally spaced points within the interval
    x = np.linspace(a, b, n+1)

    # Evaluate the function at each point
    y = f(x)

    # Calculate the trapezoid approximation of the integral
    integral = h * (np.sum(y) - 0.5 * (y[0] + y[-1]))

    # Calculate the error
    exact = exact_integral(a, b)
    error = np.abs(integral - exact)
    errors.append(error)

# Plot the graph of error vs number of intervals
plt.plot(n_values, errors)
plt.xlabel('Number of Intervals')
plt.ylabel('Error')
plt.title('Error vs Number of Intervals using Trapezoid Rule')
plt.grid(True)
plt.show()
```
